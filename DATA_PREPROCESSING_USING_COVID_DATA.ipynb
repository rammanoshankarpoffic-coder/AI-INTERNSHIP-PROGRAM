```python
import pandas as pd
from scipy.stats.mstats import winsorize

df = pd.read_csv('preprocessing_dataset.csv')


age_data = df['Age'].dropna()

print("Original Data:")
print(age_data)


winsorized_age = winsorize(age_data, limits=[0.20, 0.20])

print("\nWinsorized Data:")
print(winsorized_age)
```

```python
import numpy as np

log_age = np.log1p(age_data)

print(log_age)
```

```python
from scipy.stats import boxcox

positive_age = age_data[age_data > 0]

boxcox_data, lambda_value = boxcox(positive_age)

print(boxcox_data)
print(lambda_value)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_age = scaler.fit_transform(df[['Age']].dropna())

print(scaled_age)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

standard_age = scaler.fit_transform(df[['Age']].dropna())

print(standard_age)
```
